## Extraindo subsets

```python
colunas_csv = ['Nome', 'Estado', 'Sigla Gênero', 'Idade', 'Primeira legislatura']
df_deputados_csv = df_deputados_csv[colunas_csv]
df_deputados_csv
```

```python
df_deputados_json.columns
```

```python
colunas_json = ['nome', 'siglaPartido', 'siglaUf']
df_deputados_json[colunas_json]
```

## Unindo as bases de dados

Juntar as duas bases num `inner join` usando as colunas de nome

```python
df_deputados_final = pd.merge(
    df_deputados_json,
    df_deputados_csv,
    left_on='nome',
    right_on='Nome',
    how='inner'
)

df_deputados_final = df_deputados_final.drop(columns=['Nome'])

print(df_deputados_final.head(3))
```

## Análise


### 1. Qual a quantidade de parlamentares por estado e gênero?

```python
qtd_por_estado_genero = df_deputados_final.groupby(['Estado', 'Gênero']).size()
qtd_por_estado_genero
```

---
Tabela 1: Contagem de Parlamentares por Estado (Heatmap de Gênero)

---

```python
df_qtd_por_estado_genero = qtd_por_estado_genero.unstack(fill_value=0)

df_qtd_por_estado_genero['Total'] = df_qtd_por_estado_genero.sum(axis=1)
df_qtd_por_estado_genero = df_qtd_por_estado_genero.sort_values(by='Total', ascending=False)
df_qtd_por_estado_genero = df_qtd_por_estado_genero.drop(columns=['Total'])

df_qtd_por_estado_genero \
  .sort_values(by='Feminino', ascending=False) \
  .style \
    .background_gradient(subset=['Feminino'], cmap='PuRd') \
    .background_gradient(subset=['Masculino'], cmap='Blues')
```

---
Gráfico 1. Representação por Estado e Gênero na Câmara

---

```python
import matplotlib.pyplot as plt

CORES = {'Masculino': '#00BFFF', 'Feminino': '#FF69B4'}

fig, ax = plt.subplots(figsize=(10, 8))

df_plot = df_qtd_por_estado_genero[['Masculino', 'Feminino']]

df_plot.plot(
    kind='barh',
    stacked=True,
    color=CORES,
    ax=ax
)

masculino_valores = df_plot.iloc[:, 0]
feminino_valores = df_plot.iloc[:, 1]

deslocamento = 0.3

# para cada estado, valor masculino e valor feminino
for estado, masc, fem in zip(df_plot.index, masculino_valores, feminino_valores):

    # rótulos (masculino)
    ax.text(
        masc - deslocamento,
        df_plot.index.get_loc(estado),  # posição Y (número da linha)
        f'{masc}',
        ha='right',
        va='center',
        color='black',
        fontsize=8)

    # rótulo (feminino)
    if fem > 0:
        pos_x_feminino = masc + fem - deslocamento
        ax.text(
            pos_x_feminino,
            df_plot.index.get_loc(estado),
            f'{fem}',
            ha='right',
            va='center',
            color='black',
            fontsize=8
        )

ax.set_title('Representação por Estado e Gênero na Câmara', fontsize=14)
ax.set_xlabel('Quantidade de Parlamentares', fontsize=10)
ax.grid(axis='x', linestyle='--', alpha=0.7)
ax.invert_yaxis()

plt.tight_layout()
plt.show()
```

Qual a média de idade por gênero de parlamentares?
---

```python
media_idade_genero = df_deputados_final.groupby('Gênero')['Idade'].mean().round(1)
df_deputados_final.groupby('Gênero')['Idade'].mean()

media_idade_genero
```

---
Gráfico 2. Diagrama de caixa da Idade média dos parlamentares por Gênero

---

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))

ax = sns.boxenplot(
    data=df_deputados_final,
    x='Gênero',
    y='Idade',
    hue='Gênero',
    palette={'Masculino': '#00BFFF', 'Feminino': '#FF69B4'},
    legend=False,
    width=0.4
)

ax.text(
    0, media_idade_genero['Masculino'],
    f'{media_idade_genero['Masculino']} anos',
    ha='center',
    va='bottom',
    fontsize=10,
    color='black',)

ax.text(
    1, media_idade_genero['Feminino'],
    f'{media_idade_genero['Feminino']} anos',
    ha='center',
    va='bottom',
    fontsize=10,
    color='black',)

plt.title('Idade média dos parlamentares por Gênero')
plt.show()
```

Quais partidos têm a maior quantidade de deputadas mulheres e qual é a proporção de mulheres em relação aos homens em cada partido?
---

- DataFrame apenas com mulheres

```python
df_mulheres = df_deputados_final[df_deputados_final['Gênero'] == 'Feminino']
df_mulheres
```

---
Tabela 3. Partidos com a maior quantidade de deputadas mulheres

---

```python
mulheres_partido = df_mulheres['siglaPartido'].value_counts()
mulheres_partido.head(10)
```

---

- Criando novas colunas: Número de homens

```python
genero_partido = df_deputados_final.groupby(['siglaPartido', 'Gênero']).size().unstack(fill_value=0)
genero_partido
```

- Total de parlamentares

```python
genero_partido['Total'] = genero_partido.sum(axis=1)
genero_partido
```

- Porcentagem de mulheres (em relação ao total)

```python
genero_partido['% Mulheres'] = (genero_partido['Feminino'] / genero_partido['Total']) * 100
genero_partido
```

---
Tabela 4. Proporção de mulheres em relação aos homens em cada partido

---

```python
resultado = genero_partido.sort_values(by='Feminino', ascending=False)
resultado['% Mulheres'] = resultado['% Mulheres'].apply(lambda x: f'{x:.2f}%')
resultado
```

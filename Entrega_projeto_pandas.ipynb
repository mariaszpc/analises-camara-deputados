#### Lidando com dados em csv

Execute o bloco de código a seguir para transformar os dados csv em Data Frame

```python
# Definindo o caminho para buscar a base de dados
base_dados_deputados_csv = '/content/deputados_legislatura_57_base_tratada.csv'

# Transformando a base de dados de csv para o formato Data Frame
df_deputados_csv = pd.read_csv(base_dados_deputados_csv)

#selecionar apenas algumas colunas
colunas_csv = ['Nome', 'Estado', 'Sigla Gênero', 'Gênero', 'Data de nascimento', 'Idade', 'Primeira legislatura']
df_deputados_csv = df_deputados_csv[colunas_csv]

# Exibir as colunas do dataframe
print(df_deputados_csv.columns)

# Exibir as quantidade de linhas
df_deputados_csv.value_counts()
```

#### Lidando com dados em json

Execute o bloco de código a seguir para transformar os dados json em Data Frame


```python
import json

# Definindo o caminho para buscar a base de dados
base_dados_deputados_json = '/content/deputados_legislatura_57.json'

# Transformando a base de dados de json para o formato Data Frame
df_deputados_json = pd.read_json(base_dados_deputados_json)

#selecionar apenas algumas colunas
colunas_json = ['id', 'nome', 'siglaPartido', 'email']
df_deputados_json = df_deputados_json[colunas_json]

# Exibir as colunas do dataframe
print(df_deputados_json.columns)

# Exibir as quantidade de linhas
print(df_deputados_json.value_counts)

# Imprime os as colunas do dataframe
print(df_deputados_json.columns)
print(df_deputados_json.value_counts)
```

#### Juntar bases de dados em uma só

Vamos utilizar o conceito de [joins (ou junção)](https://https://www.alura.com.br/artigos/join-e-seus-tipos?srsltid=AfmBOorqiWzXLMyZKkskHWph08lQWzlGAOAJJ-DQR-cC4DF46cQ_VnA1 ) para mesclar as duas bases em uma só a partir da coluna "nome"

```python
# Juntar as duas bases num inner join usando as colunas de nome
df_deputados_final = pd.merge(
    df_deputados_json,
    df_deputados_csv,
    left_on='nome',
    right_on='Nome',
    how='inner'
)

# Remover a coluna duplicada Nome
df_deputados_final = df_deputados_final.drop(columns=['Nome'])

print(df_deputados_final.head(3))
```

Qual a quantidade de parlamentares por estado e gênero?
---

```python
qtd_por_estado_genero = df_deputados_final.groupby(['Estado', 'Gênero']).size()
qtd_por_estado_genero
```

---
Tabela 1: Contagem de Parlamentares por Estado (Heatmap de Gênero)

---

```python
df_qtd_por_estado_genero = qtd_por_estado_genero.unstack(fill_value=0)

df_qtd_por_estado_genero['Total'] = df_qtd_por_estado_genero.sum(axis=1)
df_qtd_por_estado_genero = df_qtd_por_estado_genero.sort_values(by='Total', ascending=False)
df_qtd_por_estado_genero = df_qtd_por_estado_genero.drop(columns=['Total'])

df_qtd_por_estado_genero \
  .sort_values(by='Feminino', ascending=False) \
  .style \
    .background_gradient(subset=['Feminino'], cmap='PuRd') \
    .background_gradient(subset=['Masculino'], cmap='Blues')
```

---
Gráfico 1. Representação por Estado e Gênero na Câmara

---

```python
import matplotlib.pyplot as plt

CORES = {'Masculino': '#00BFFF', 'Feminino': '#FF69B4'}

fig, ax = plt.subplots(figsize=(10, 8))

df_plot = df_qtd_por_estado_genero[['Masculino', 'Feminino']]

df_plot.plot(
    kind='barh',
    stacked=True,
    color=CORES,
    ax=ax
)

masculino_valores = df_plot.iloc[:, 0]
feminino_valores = df_plot.iloc[:, 1]

deslocamento = 0.3

# para cada estado, valor masculino e valor feminino
for estado, masc, fem in zip(df_plot.index, masculino_valores, feminino_valores):

    # rótulos (masculino)
    ax.text(
        masc - deslocamento,
        df_plot.index.get_loc(estado),  # posição Y (número da linha)
        f'{masc}',
        ha='right',
        va='center',
        color='black',
        fontsize=8)

    # rótulo (feminino)
    if fem > 0:
        pos_x_feminino = masc + fem - deslocamento
        ax.text(
            pos_x_feminino,
            df_plot.index.get_loc(estado),
            f'{fem}',
            ha='right',
            va='center',
            color='black',
            fontsize=8
        )

ax.set_title('Representação por Estado e Gênero na Câmara', fontsize=14)
ax.set_xlabel('Quantidade de Parlamentares', fontsize=10)
ax.grid(axis='x', linestyle='--', alpha=0.7)
ax.invert_yaxis()

plt.tight_layout()
plt.show()
```

- Criando novas colunas: Número de homens

```python
genero_partido = df_deputados_final.groupby(['siglaPartido', 'Gênero']).size().unstack(fill_value=0)
genero_partido
```

```python
genero_partido['Total'] = genero_partido.sum(axis=1)
genero_partido
```

- Porcentagem de mulheres (em relação ao total)

```python
genero_partido['% Mulheres'] = (genero_partido['Feminino'] / genero_partido['Total']) * 100
genero_partido
```
